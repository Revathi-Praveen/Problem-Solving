#### <strong>PROBLEM 1.</strong> A news article discusses the number of promises made by political candidates and the number of promises kept after they are elected. The table below shows the promises made by the candidates X and the promises Y each kept. Create a scatter plot with a linear regression line. Calculate the correlation coefficient. What is the relationship between promises made and promised kept?

<table>
    <tr>
        <td><strong>Promises made by candidates, X</strong></td>
        <td>20</td><td>30</td><td>30</td><td>40</td><td>50</td><td>50</td><td>60</td>
    </tr>
    <tr>
        <td align="right"><strong>Promises kept, Y</strong></td>
        <td>7</td><td>6</td><td>5</td><td>4</td><td>3</td><td>2</td><td>1</td>
    </tr>
</table>

```python
import matplotlib.pyplot as plt
import numpy as np
X=[20,30,30,40,50,50,60]
Y=[7,6,5,4,3,2,1]

#Plotting least square regression line
mean_x = np.mean(X)
mean_y = np.mean(Y)
 
n = len(X)

# Using the formula to calculate 'm' and 'c'
numer = 0
denom = 0
for i in range(n):
    numer += (X[i] - mean_x) * (Y[i] - mean_y)
    denom += (X[i] - mean_x) ** 2
    m = numer / denom
    c = mean_y - (m * mean_x)

# Printing coefficients
print("Slope and intercept")
print(f'm={round(m,2)}, c={round(c,2)}')

# Printing correlation coefficient
print(f"\nCorrelation coefficient: {round(np.corrcoef(X,Y)[0][1],2)}")

# Plotting Values and Regression Line
 
max_x = np.max(X)
min_x = np.min(X)
 
# Calculating line values x and y
x = np.linspace(min_x, max_x, 1000)
y = c + m * x
 
# Ploting Line
plt.plot(x, y, color='#58b970', label='Regression Line')

# Ploting Scatter Points
plt.scatter(X, Y, c='#ef5423', label='Scatter Plot')
 
plt.xlabel('Promises made')
plt.ylabel('Promises kept')
plt.legend()
plt.title(f'Least square regression line y={round(m,2)}x + {round(c,2)}')
plt.show()

```

###### Intuition from relationship:
 - There is a strong LINEAR relationship between promises made and promises kept in negative direction (r close to -1)
 - There is a very strong negative correlation i.e., as promises made increase, the promises kept decreases (inverse relationship)
 - However, that doesn't imply causation, i.e., increase in number of promises made doesn't imply promises kept

#### <strong>PROBLEM 3.</strong> SJSU bought 36 bulbs for its classroom projectors. The manufacturer claimed that the bulbs should each last more than 800 hours. After using the bulbs, the university discovered that their mean lifetime was 816 hours with a standard deviation of 70 hours. At the 5% level of significance, should the university accept the manufacturer's claim?

* Null hypothesis: Bulb lasts more than 800 hours (mu>800)
* Alternate hypothesis: Bulb lasts equals or less than 800 hours (mu<=800)
    

#### why t tset??

###### Though n>30 & population mean is known, we should go with t- test because of the below reasons:
* Population standard deviation is unknown (manufacturer hasn't claimed any standard deviation)
https://www.analyticsvidhya.com/blog/2020/06/statistics-analytics-hypothesis-testing-z-test-t-test/

###### What test?
* It is a one tailed t test (upper tailed)



```python
mu_population=800

# Number of samples bought from the manufacturer
n=36
x_sample=816
stdev_sample=70

# since it is one tailed t test with degree of freedom of 36-1=35
t_critical=1.689

# t score calculation
t=(x_sample-mu_population)/(stdev_sample/(n)**(1/2))
print(f"The t-score is: {round(t,3)}")
print(f"The t-critical is: {round(t_critical,3)}")
```

###### Inference

* Since, the t score is lesser than the critical value corresponding to 5% level of significance, 
* we will not reject the null hypothesis

* So, we can accept the manufacturer's claim

#### <strong>PROBLEM 4.</strong> A widget making machine is supposed to make widgets that are 0.05 inches thick. Ten widgets are tested each day to ensure the machine is properly calibrated. Today's test averaged 0.053 inches thick with a standard deviation of 0.003. At the 5% level of significance, is the machine properly calibrated?

* Null hypothesis: Widgets are 0.05 inches thick (mu=0.05)
* Alternate hypothesis: Widgets are not 0.05 inches thick (mu!=0.05)

#### why t tset??


###### Though population mean is known, we should go with t- test because of the below reasons:
* Population standard deviation is unknown (nothing is told about the standard deviation from all the widgets from the widget manufacturing company) 
* n<30 (n=10)

###### What test?
* It is a two tailed t test



```python
mu_population=0.05

# Number of samples bought from the manufacturer
n=10
x_sample=0.053
stdev_sample=0.003

# since it is two tailed t test with degree of freedom of 10-1=9
t_critical=2.262

# t score calculation
t=(x_sample-mu_population)/(stdev_sample/(n)**(1/2))
print(f"The t-score is: {round(t,3)}")
print(f"The t-critical is: {round(t_critical,3)}")
```

###### Inference

* Since, the t score is greater than the critical value corresponding to 5% level of significance, 
* we will reject the null hypothesis

* So, the machines aren't calibrated properly at 5% level of significance

```python
mu_population=380

# Number of samples bought from the manufacturer
n=50
x_sample=390
stdev_sample=35.2

# since it is two tailed t test with degree of freedom of 50-1=49
t_critical=1.676

# t score calculation
t=(x_sample-mu_population)/(stdev_sample/(n)**(1/2))
print(f"The t-score is: {round(t,4)}")
print(f"The t-critical is: {round(t_critical,3)}")

print(f"The p-value is: {0.025}")

# Use the below p-value calculator for t value of 2.0088 and dof of 49
# https://www.socscistatistics.com/pvalues/tdistribution.aspx
```

#### <strong>PROBLEM 7.</strong> Use the data from Problem 6. Calculate and print:
- the variation in the pounds lost that cannot be explained by the weeks on the diet
- the total variation in the pounds lost
- the percentage of variation that is explained by the weeks on the diet
- the correlation coefficient 

```python
# the variation in the pounds lost that cannot be explained by the weeks on the diet

# This is the SUM OF SQUARES OF RESIDUALS
import statsmodels.api as sm
import pandas as pd
import numpy as np

#define response variable
y = Y

#define predictor variables
x = X

#add constant to predictor variables
x = sm.add_constant(x)

#fit linear regression model
model = sm.OLS(y, x).fit()

ss_residual=np.sum((model.fittedvalues - y)**2)
print(f"Answer 1) Unexplained variation: {round(ss_residual,2)}")
ss_regression = np.sum((model.fittedvalues - sum(y)/len(y))**2)
print(f"(Explained variation: {round(ss_regression,2)})")


# the total variation in the pounds lost
ss_total=ss_regression+ss_residual
print(f"\nAnswer 2) Total variation in pounds lost: {round(ss_total,2)}")

# the percentage of variation that is explained by the weeks on the diet is r^2
print(f"\nAnswer 3) The percentage of variation that is explained by the weeks on the diet: {round(np.corrcoef(X,Y)[0][1]**2*100,1)}%")


# the correlation coefficient
print(f"Answer 4) Correlation coefficient: {round(np.corrcoef(X,Y)[0][1],3)}")
```

```python
# Validating the results: r squared is explained variation as a ratio to total sum of squared errors
r_squared=(ss_total-ss_residual)/ss_total
r=r_squared**(1/2)
print(f"Correlation coefficient: {round(r,2)}")
print("BOTH THE CORRELATION COEFFICIENT VALUES MATCH, SO ERROR VALUES ARE RIGHT!")
```

```python
import statistics as stat
import matplotlib.pyplot as plt
from scipy import stats

def show_time_lines(title, x_label, y_label, x_values, y_values, 
                    regression='yes', mv_amt_1=0, mv_amt_2=0, 
                    alpha_1=0, alpha_2=0, 
                    size=()):
    """
    @param title the chart title.
    @param x_label the x-axis label.
    @param y_label the y-axis label.
    @param x_values the independent x values to plot.
    @param y_values the dependent y values to plot.
    """
    if size != ():
        figure = plt.figure(figsize=size)

    last = []
    plt.xticks(x_values)
        
    # First show the data line.
    plt.plot(x_values, y_values, linewidth=5)
    
    last_1 = []
    last_2 = []
    linear_line   = []
    mv_avg_line_1 = []
    mv_avg_line_2 = []
    smooth_line_1 = []
    smooth_line_2 = []

    if regression == 'yes':
        linear_regression = stats.linregress(x=x_values,
                                             y=y_values)
        m = linear_regression.slope
        b = linear_regression.intercept

    
    # Calculate the least squares trend line, 
    # the simple moving average line,
    # and the exponentially smoothed line.
    for i in range(len(x_values)):
        x = x_values[i]
        y = y_values[i]
        
        # Linear line
        if regression == 'yes': 
            y_hat = m*x + b
            linear_line.append(y_hat)
        
        # Moving average line #1
        if mv_amt_1 != 0:
            last_1.append(y)
            
            if len(last_1) == mv_amt_1:
                mv_avg_line_1.append(stat.mean(last_1))
                last_1.pop(0)
        
        # Moving average line #2
        if mv_amt_2 != 0:
            last_2.append(y)
            
            if len(last_2) == mv_amt_2:
                mv_avg_line_2.append(stat.mean(last_2))
                last_2.pop(0)

        # Exponentially smoothed line #1
        if alpha_1 != 0:
            if i == 0:
                y_smooth_1 = y
            else:
                y_smooth_1 = alpha_1*y + (1 - alpha_1)*y_smooth_1
                
            smooth_line_1.append(y_smooth_1)
                 
        # Exponentially smoothed line #2
        if alpha_2 != 0:
            if i == 0:
                y_smooth_2 = y
            else:
                y_smooth_2 = alpha_2*y + (1 - alpha_2)*y_smooth_2
                
            smooth_line_2.append(y_smooth_2)

    if regression == 'yes':
        plt.plot(xs, linear_line, color='red')
    
    if mv_amt_1 != 0:
        plt.plot(xs[mv_amt_1 - 1:], mv_avg_line_1, color='lightgreen')
    
    if mv_amt_2 != 0:
        plt.plot(xs[mv_amt_2 - 1:], mv_avg_line_2, color='green')
        
    if alpha_1 != 0:
        plt.plot(xs, smooth_line_1, color='gray', linestyle='--')
        
    if alpha_2 != 0:
        plt.plot(xs, smooth_line_2, color='black', linestyle='--')

    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)

    plt.show()
```

```python
xs = list(range(1993, 2019))
ys = [9.9,22.2,11.4,14.8,19.7,14.9,15.9,13.4,12.0,7.9,12.9,16.8,11.6,14.9,13.3,20.2,14.3,20.4,13.0,10.0,15.9,17.4,18.7,14.3,16.2,21.1]

show_time_lines('Snowfall over years', 'Years', 'Snowfall', xs, ys,size=(20,10))
```
