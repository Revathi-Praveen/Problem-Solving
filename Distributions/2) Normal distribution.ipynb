## Normal / Gaussian distribution

###### Import all the necessary external packages

```python
import random as rnd
import numpy as np
import seaborn as sns
```

###### Setting the mean and sigma

```python
mean=12.5 
sigma=1
```

###### Create a list of 1000 random numbers from 0 to 25 (inclusive)

```python
#Setting initial values so that the pseudo random values don't change 
rnd.seed(0)
normal_samples=[int((rnd.gauss(mean,sigma))) for i in range(0,1000)]
```

```python
# Finding the frequency of each values of the list of random numbers
value,frequency=np.unique(normal_samples,return_counts=True)
```

```python
print((min(frequency),max(frequency)),(max(value),max(value)))
```

###### Plotting the gaussian distribution with 1000 samples, mean=12.5, sigma=2

```python
a=sns.lineplot(x=value,y=frequency)
a.set_xlim((mean-mean,mean+mean))
sns.set_theme(font_scale=1)
a.set_title('Normal distribution (sample size: 1000)',fontsize='15',color='brown')
# a.set_ylim((0,max(frequency)+min(frequency)))
# a.set_xlim((0,max(value)))
a.set_xlabel('Random Values',fontsize='12',color='brown')
a.set_ylabel('Frequency',fontsize='12',color='brown')
```

```python
#Observation
'''Though the distribution looks like a normal distribution, it is not smooth. 
So, we need to increase the sample size'''

''' Change the sigma value to 2 and see the curve smoothen and getting wider'''
```

###### Increasing the sample size to 10 thousand

```python
sigma=2
```

```python
rnd.seed(0)
normal_samples_large=[int((rnd.gauss(mean,sigma))) for i in range(0,10_000)]
value_large,frequency_large=np.unique(normal_samples_large,return_counts=True)

a=sns.lineplot(x=value_large,y=frequency_large)
a.set_xlim((mean-mean,mean+mean))
sns.set_theme(font_scale=1)
a.set_title('Normal distribution (sample size: 10,000)',fontsize='15',color='brown')
# a.set_xlim((0,max(value)))
a.set_xlabel('Random Values',fontsize='12',color='brown')
a.set_ylabel('Frequency',fontsize='12',color='brown')
```

###### Increasing the sample size to 10 million

```python
#Try experimenting this value to see how the normal distribution varies

sigma=3
```

```python
rnd.seed(0)
normal_samples_very_large=[int((rnd.gauss(mean,sigma))) for i in range(0,10_000_000)]
```

```python
value_very_large,frequency_very_large=np.unique(normal_samples_very_large,return_counts=True)
```

```python
a=sns.lineplot(x=value_very_large,y=frequency_very_large,color='brown')
sns.set_theme(font_scale=1)
a.set_title('Normal distribution (sample size: 10 million)',fontsize='15',color='brown')
a.set_xlim((min(value_very_large),max(value_very_large)))
a.set_xlabel('Random Values',fontsize='12',color='brown')
a.set_ylabel('Frequency',fontsize='12',color='brown')
```

###### Values outside the range is ignored using if statement

```python
if min(value_very_large)<0 or max(value_very_large)>25:
    a=sns.lineplot(x=value_very_large,y=frequency_very_large,color='brown')
    a.set_xlim((mean-mean,mean+mean))
    a.set_title('Normal distribution (sample size: 10 million)',fontsize='15',color='brown')
    a.set_xlabel('Random Values',fontsize='12',color='brown')
    a.set_ylabel('Frequency',fontsize='12',color='brown')
```

######  Using a different method to eliminate the values of the gaussian distribution outside the range 0-25  and using 1M as a sample size
###### and experimenting with sigma value (sigma=3) 

```python
mean=12.5
sigma=3.5
```

```python
rnd.seed(0)
normal_sample=[]
for i in range(0,1_000_000):
    gauss_random=int(rnd.gauss(mean,sigma))
    if gauss_random>=0 and gauss_random<=25:
        normal_sample.append(gauss_random)
```

```python
#Check if the elements outside the 0 to 25 range is eliminated
print([x for x in normal_sample if x>25 or x<0])
print(len(normal_sample))
```

```python
#Get the frequency to plot the normal distribution
value,frequency=np.unique(normal_sample,return_counts=True)
```

```python
#Plot the graph
a=sns.lineplot(x=value,y=frequency,color='brown')
sns.set_theme(font_scale=1)
a.set_title('Normal distribution (sample size: 1 million)',fontsize='15',color='brown')
a.set_xlim((min(value),max(value)))
a.set_xlabel('Random Values',fontsize='12',color='brown')
a.set_ylabel('Frequency',fontsize='12',color='brown')
```

###### Takeaway

```python
'''Increasing the sample size makes the normal distribution smoother

Increasing the standard deviation value makes the distribution wider'''
```
