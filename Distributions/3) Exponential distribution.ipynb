## Exponential distribution

###### Import all the necessary external packages

```python
import random as rnd
import numpy as np
import seaborn as sns
```

###### Assigning mean and lambda

```python
mean=12.5
lambd=1/mean
```

###### Create a list of 1000 random numbers from 0 to 25 (inclusive)

```python
rnd.seed(0)
exponential_dist=[int(rnd.expovariate(lambd=lambd)) for i in range(0,1000)]
len(exponential_dist)
```

###### Finding the frequency of each values of the list of random numbers

```python
value,frequency=np.unique(exponential_dist,return_counts=True)
```

###### Plotting the exponential distribution

```python
sns.set_style('whitegrid')
a=sns.lineplot(x=value,y=frequency,color='brown')
sns.set_theme(font_scale=1)
a.set_title('Exponential distribution (sample size: 1000)',fontsize='15',color='brown')
a.set_ylim((0,max(frequency)+min(frequency)))
a.set_xlim((0,max(value)))
a.set_xlabel('Random Values',fontsize='12',color='brown')
a.set_ylabel('Frequency',fontsize='12',color='brown')
```

```python
#Observation
'''The exponential graph needs smoothening'''
```

###### Changing the sign of lambda reverses the exponential distribution

```python
lambd=-1/mean

exponential_dist=[int(rnd.expovariate(lambd=lambd)) for i in range(0,1000)]

value,frequency=np.unique(exponential_dist,return_counts=True)

sns.set_style('whitegrid')
a=sns.lineplot(x=value,y=frequency,color='brown')
sns.set_theme(font_scale=1)
a.set_title('Exponential distribution (sample size: 1000)',fontsize='15',color='brown')
# a.set_ylim((0,max(frequency)+min(frequency)))
# a.set_xlim((0,max(value)))
a.set_xlabel('Random Values',fontsize='12',color='brown')
a.set_ylabel('Frequency',fontsize='12',color='brown')
```

###### Increasing the sample size to 10 million

```python
mean=12.5
lambd=1/mean

exponential_dist_large=[int(rnd.expovariate(lambd=lambd)) for i in range(0,10_000_000)]

value_large,frequency_large=np.unique(exponential_dist_large,return_counts=True)

sns.set_style('whitegrid')
a=sns.lineplot(x=value_large,y=frequency_large,color='brown')
sns.set_theme(font_scale=1)
a.set_title('Exponential distribution (sample size: 10 Million)',fontsize='15',color='brown')
a.set_ylim((0,max(frequency_large)+min(frequency_large)))
a.set_xlim((0,max(value_large)))
a.set_xlabel('Random Values',fontsize='12',color='brown')
a.set_ylabel('Frequency',fontsize='12',color='brown')
```

###### Takeaway
